```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50)
x_pca=pca.fit_transform(x_scaled)
```

```python
pca.explained_variance_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y.values.ravel(),test_size=0.2,random_state=42)
```

```python
y_train=np.where(y_train==-1,0,y_train)
y_test=np.where(y_test==-1,0,y_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

models={
  'SVC': Pipeline([
    ('scaler',StandardScaler()),
    ('model',SVC())
  ]),
  'KNN': Pipeline([
    ('scaler',StandardScaler()),
    ('model',KNeighborsClassifier())
  ]),
  'XgBoost': Pipeline([
    ('scaler',StandardScaler()),
    ('model',XGBClassifier())
  ])
}
```

```python
result={}
for name,pipeline in models.items():
  model=pipeline.fit(x_train,y_train)
  y_pred=model.predict(x_test)
 
  result[name]={
    'Accuracy': accuracy_score(y_test,y_pred),
    'Precision': precision_score(y_test,y_pred,average='weighted'),
    'Recall': recall_score(y_test,y_pred,average='weighted'),
    'f1': f1_score(y_test,y_pred,average='weighted')
  }

result_df=pd.DataFrame(result).T
result_df.sort_values(by='Accuracy',ascending=False)
```

```python
from sklearn.model_selection import GridSearchCV

pipeline=Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(class_weight='balanced'))
])

param_grid={
    'svc__C': [0.01, 0.1, 1, 10, 100],
    'svc__gamma': [0.001, 0.01, 0.1, 1],
    'svc__kernel': ['linear', 'rbf']
}

grid=GridSearchCV(pipeline,param_grid=param_grid)
grid.fit(x_train,y_train)
```

```python
print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
y_pred = grid.best_estimator_.predict(x_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
y_pred_full=grid.best_estimator_.predict(x_scaled)

pca_vis=PCA(n_components=2)
x_pca_vis=pca.fit_transform(x_scaled)

plt.scatter(x_pca_vis[:,0],x_pca_vis[:,1],c=y_pred_full,cmap='plasma',alpha=0.7)
plt.title('Predicted Labels in PCA (2D) Space')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Predicted class')
plt.show()
```

```python
pca1=PCA()
x_pca1=pca1.fit_transform(x_scaled)

cum_var=pca1.explained_variance_ratio_.cumsum()

plt.plot(range(1,len(cum_var)+1),cum_var,marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Explained Variance vs Components")
plt.grid(True)
plt.show()
```

```python
n_components=np.argmax(cum_var>=0.95)+1
print("Number of components for 95% variance:",n_components)
```

```python
pca_final=PCA(n_components=n_components)
x_pca_final=pca_final.fit_transform(x_scaled)
```

```python
print("Original shape (before PCA):", x_scaled.shape)
print("Reduced shape (after PCA):", x_pca_final.shape)
print("Number of PCA components used:", x_pca_final.shape[1])
```

```python
components=pd.DataFrame(pca.components_, columns=x_train.columns)

top_features_per_pc={
    f"PC{i+1}":row.abs().nlargest(5).index.tolist()
    for i, row in components.iterrows()
}

for pc, features in top_features_per_pc.items():
    print(f"{pc}: {features}")
```
