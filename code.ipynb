```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y=pd.DataFrame(y)
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y.values.ravel(),test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca_full=PCA()
x_pca_full=pca_full.fit_transform(x_train_scaled)

cum_var=pca_full.explained_variance_ratio_.cumsum()

plt.plot(range(1,len(cum_var)+1),cum_var,marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Explained Variance vs Components")
plt.grid(True)
plt.show()
```

```python
n_components=np.argmax(cum_var>=0.95)+1
print("Number of components for 95% variance:",n_components)
```

```python
pca=PCA(n_components=n_components)
x_train_pca=pca.fit_transform(x_train_scaled)
x_test_pca=pca.transform(x_test_scaled)
```

```python
print("Original shape (before PCA):", x_train_scaled.shape)
print("Reduced shape (after PCA):", x_train_pca.shape)
print("Number of PCA components used:", x_train_pca.shape[1])
```

```python
components=pd.DataFrame(pca.components_, columns=x_train.columns)

top_features_per_pc={
    f"PC{i+1}":row.abs().nlargest(10).index.tolist()
    for i, row in components.iterrows()
}

for pc, features in top_features_per_pc.items():
    print(f"{pc}: {features}")
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

models={
  'SVC': SVC(),
  'KNN': KNeighborsClassifier(),
  'Random Forest': RandomForestClassifier(),
  'XgBoost': XGBClassifier()
}
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

result={}
for name,models in models.items():
  model=models.fit(x_train_pca,y_train)
  y_pred=model.predict(x_test_pca)
 
  result[name]={
    'Accuracy': accuracy_score(y_test,y_pred),
    'Precision': precision_score(y_test,y_pred,average='weighted'),
    'Recall': recall_score(y_test,y_pred,average='weighted'),
    'f1': f1_score(y_test,y_pred,average='weighted')
  }

result_df=pd.DataFrame(result).T
result_df.sort_values(by='Accuracy',ascending=False)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

xgb = XGBClassifier()

grid = GridSearchCV(estimator=xgb,param_grid=param_grid,cv=5)
grid.fit(x_train_pca,y_train)
```

```python
print(f'Best Parameters: {grid.best_params_}')
print(f'Best score: {grid.best_score_}')

y_pred_grid=grid.predict(x_test_pca)
print(f'Accuracy: {accuracy_score(y_test,y_pred_grid)}')
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_grid))
```

```python
pca_vis=PCA(n_components=2)
x_pca_vis=pca_vis.fit_transform(x_train_scaled)

y_pred_full=grid.best_estimator_.predict(x_train_pca)
plt.scatter(x_pca_vis[:,0],x_pca_vis[:,1],c=y_pred_full,cmap='plasma',alpha=0.7)
plt.title('Predicted Labels in PCA (2D) Space')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Class')
plt.show()
```
